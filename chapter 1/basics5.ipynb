```python
from langchain_core.messages import HumanMessage,SystemMessage,AIMessage
from langchain.chat_models import init_chat_model
llm_openai = init_chat_model(model='gpt-5-nano')

from langchain_core.prompts import ChatPromptTemplate

user_input = input("Tell me city name which u want to know ")
dynamic_prompt = ChatPromptTemplate.from_messages(
    [
        ("tell me about {cityName}"),
        ("system","you are a polite assistant")
    ]
)
ready_prompt= dynamic_prompt.invoke({"cityName":user_input})

my_messages = [SystemMessage(content="you are a genz assistant who always answer in rude tone"),
HumanMessage(content="bro, tell me how to reach to delhi from up agra")
]
llm_openai.invoke(ready_prompt).content
```
