# **Pydent Model**

```python
from langchain_core.messages import HumanMessage,SystemMessage,AIMessage
from langchain.chat_models import init_chat_model
llm_openai = init_chat_model(model='gpt-5-nano')                
```

```python
from pydantic import BaseModel, Field

class llm_schema(BaseModel):
    productKey:str=Field(description="The product unique key")
    description:str=Field(description="The description of product")
    
```

```python
llm_structured_output = llm_openai.with_structured_output(llm_schema)
llm_structured_output.invoke("tell me about rubber band")
```
